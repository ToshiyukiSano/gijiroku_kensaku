ファイル全体を読み込む

```python
# input_text_long = 

with open(file_name, 'r', encoding="utf-8") as input_file:
    input_text_long = input_file.read()

input_text_long
```

```python
# 形態素解析ライブラリ janome のインポート

from janome.tokenizer import Tokenizer
```

```python
# どっかからコピペした関数

def tokenize(text):
    t = Tokenizer()
    tokens = t.tokenize(text.replace('・',' '))
    return [x.surface.strip() for x in tokens]
```

```python
tokens = tokenize(input_text_long)
tokens
```

```python
# 並べ替え用のライブラリのインポート
#　仕様は　https://docs.python.org/ja/3/library/collections.html?highlight=collections%20counter#collections.Counter

import collections
```

```python
count = collections.Counter(tokens)
print(count)
```

```python
type(count)
```

出現頻度上位のキーワード、出現回数を出力する。

```python
# tokensの中から、出現頻度の高い順にnum個並べる

num = 20

collections.Counter(tokens).most_common(num)
```
