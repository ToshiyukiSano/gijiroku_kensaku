ファイルを読み込み、input_text[]に１行ずつ格納する。

```python
input_text = {}

with  open('hiraya_whisper_medium.txt', encoding="utf-8") as input_file:
    for i, line in enumerate(input_file):
        #print(i," ",line)
        input_text[i] = line

for i in input_text:
    print(i," ", input_text[i])

print(len(input_text))

```

格納したデータの各行にキーワードがあるかどうか確認し、ある行の番号をkeyword_line[]に格納する。


```python
search_text = "施工" ############ 検索する言葉 ###########


keyword_line = []

for i in input_text:
    this_line = input_text[i]
    if search_text in this_line:
        print(i, " ", this_line)
        keyword_line.append(i)

print(keyword_line)
```

キーワードがある前後の数十行（zengo_linesで指定）を、決める。

```python
zengo_lines = 10  ##### キーワードがある行の前後何行分かを指定する。

# キーワードがある領域を調べ、keyword_area 配列に０，１で入れる。
# 0: キーワードなしエリア、1:キーワードありエリア


# キーワードがある領域を示す配列の初期化　
keyword_area = [0 for i in enumerate(input_text)]
print(len(keyword_area))
print(len(input_text))

# キーワードがある領域を１にする。
# キーワードがある領域は、キーワードがある行の前後zengo_lines行とする。
for i in range(len(keyword_line)):
    print(i, " ",keyword_line[i])
    for j in range(max(0, keyword_line[i]-zengo_lines),min(len(input_text), keyword_line[i]+zengo_lines)):
        keyword_area[j] = 1

print(keyword_area)

```

キーワード領域のテキストを結合し、キーワード前後の文脈がわかる文章を作成する。

```python
keyword_sentence = []

print(len(keyword_area))

sentence_end = 0
sentence_start = 1
text_num = 0

text = ""



for i in range(len(keyword_area)):
# for i in range(10):
    if keyword_area[i] == 1: # キーワード領域であるならば
        # text  += input_text[i]
        # print(text)
        if sentence_start == 1:  # キーワード領域の最初の文であるならば
            keyword_sentence.append("...") # 
            sentence_start = 0
        keyword_sentence[text_num] += input_text[i]
    else:
        if sentence_start == 0:
            keyword_sentence[text_num] += "..."
            text_num += 1
        sentence_start = 1 
        # print(".")
```

```python
for i in range(len(keyword_sentence)):
    print(i, " ", keyword_sentence[i])
```
